```python
#6. Train
from ultralytics import YOLO
import torch

# Check if GPU is actually ready now
print(f"Is GPU available? {torch.cuda.is_available()}")

# Load the YOLO26 Nano Segmentation model
model = YOLO('yolo26n-seg.pt')

# Start Training
# I removed 'device=0' so it auto-detects the GPU we just turned on
model.train(
    data=f"{dataset.location}/data.yaml",
    epochs=120,
    imgsz=640,
    batch=16,
    project='/content/drive/MyDrive/omnivision',
    name='omni_goal_26',
    exist_ok=True
)
```
